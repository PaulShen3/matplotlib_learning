# Matplotlib散点图

语法格式：

plt.scatter(x,y,s=None,c=None,marker=None,cmap=None,norm=None,vmax=None,alpha=None,linewidths=None,*,edgecolors=None,plotnonfinite=False,data=None,**kwargs)

x,y:长度相同的数组

s:点的大小，也可以是数组，数组的每个参数为对应点的大小

c:点的颜色，默认为b

marker:点的样式，默认是‘o’

cmap:colormap，只有c是一个浮点数数组时才使用，没有申明就是image.cmap,cmap可以用不同的颜色条，如cmap = 'afmhot_r'和'viridis'

norm:默认为none，数据量都在0-1之间，只有c是一个浮点数的数组才使用

vmin，vmax：亮度设置，norm参数存在时会忽略

alpha:透明度设置，0-1之间

linewidth：标记点的长度

edgecolors：颜色或颜色序列，默认为‘face’

plotnonfinite:布尔值

**kwargs:其他参数

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([1,2,3,4,5,6,7,8])
y = np.array([1,4,9,16,7,11,23,18])
plt.scatter(x,y)
plt.show()
```

```python
x = np.array([1,2,3,4,5,6,7,8])
y = np.array([1,4,9,16,7,11,23,18])
size = [20,50,100,200,500,1000,60,90]
plt.scatter(x,y,s = size)
plt.show()
```

```python
x = np.array([1,2,3,4,5,6,7,8])
y = np.array([1,4,9,16,7,11,23,18])
size = [20,50,100,200,500,1000,60,90]
color = np.array(['red','green','black','orange','purple','beige','cyan','magenta'])
plt.scatter(x,y,s = size,c=color)
plt.show()
```

```python
x1 = np.array([1,2,3,4,5,6,7,8])
y1 = np.array([1,4,9,16,7,11,123,18])
plt.scatter(x1,y1,c='hotpink')

x2 = np.array([2,3,4,6,2,12,5,7])
y2 = np.array([1,5,7,12,63,123,51,11])
plt.scatter(x2,y2,c = 'green')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

#设置随机种子
np.random.seed(19680801)

#设置数据
n = 50
x = np.random.rand(n)
y = np.random.rand(n)

#设置散点大小
area = (30 * np.random.rand(n))**2

#设置散点颜色
color = np.random.rand(n)   #读取颜色条

#绘制散点图
plt.scatter(x,y,s = area,c = color,alpha = 0.5,cmap ='afmhot_r' )
plt.colorbar()
plt.title("RUNOOB Scatter Test")

#打印图表
plt.show()
```
