```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([1,3,4,5,8,9,6,1,3,4,5,2,4])
plt.plot(y,color = 'red',marker = 'o')
plt.show()
```

```python
#fmt参数：fmt = '[marker][line][color]'
y = np.array([6,2,13,10])
plt.plot(y,'o:r')   #o表示实心圆，:表示虚线，r表示红色
plt.show()
```

```python
#ms定义标记的大小
#mfc定义标记内部的颜色
#mec定义标记边框的颜色
```

```python
y = np.array([6,2,13,10])
plt.plot(y,marker = 'o',ms = 10)
plt.show()
```

```python
y = np.array([6,2,13,10])
plt.plot(y,marker='o',ms=20,mec='r')
plt.show()
```

```python
y = np.array([6,2,13,10])
plt.plot(y,marker = 'o',ms = 20, mec = 'r',mfc = 'g')
plt.show()
```

```python
y = np.array([6,2,13,10])
plt.plot(y,marker = 'o',ms = 20, mec = '#4CAF50',mfc = '#4CAF50') #自定义颜色
plt.show()
```

```python
y = np.array([1,3,4,5,8,9,6,1,3,4,5,2,4])
plt.plot(y,marker = 'o',ms = 10, mec = '#4CAF50',mfc = '#4CAF50')
plt.show()
```
