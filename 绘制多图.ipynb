# 可以使用subplot()和subplots()方法来绘制多个子图
subplot()在绘图时需要指定位置
subplots()可以一次生成多个，在调用时只需要调用生成对象的ax即可

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#plot1
x1 = np.array([0,6])
y1 = np.array([0,100])
plt.subplot(1,2,1)
plt.plot(x1,y1)
plt.title("plot 1")
#plot2
x2 = np.array([1,2,3,4])
y2 = np.array([1,4,9,16])
plt.subplot(1,2,2)
plt.plot(x2,y2)
plt.title("plot 2")
#给一个主title
plt.suptitle("RUNOOB subplot Test")   
plt.show()
```

```python
#引入需要的包
import numpy as np
import matplotlib.pyplot as plt

#图1
x1 = np.array([0,6])
y1 = np.array([0,100])
plt.subplot(2,2,1)
plt.plot(x1,y1)
plt.title("plot 1")

#图2
x2 = np.array([1,2,3,4])
y2 = np.array([1,4,9,16])
plt.subplot(2,2,2)
plt.plot(x2,y2)
plt.title("plot 2")

#图3
x3 = np.array([1,2,3,4])
y3 = np.array([3,5,7,9])
plt.subplot(2,2,3)
plt.plot(x3,y3)
plt.title("plot 3")

#图4
x4 = np.array([1,2,3,4])
y4 = np.array([4,5,6,7])
plt.subplot(2,2,4)
plt.plot(x4,y4)
plt.title("plot 4")

#给一个主标题并打印图画
plt.suptitle("RUNOOB subplot Test")
plt.show()
```

# subplots()使用

plt.subplots(nrows=1,ncols=1,*,sharex=False,sharey=False,squeeze=True,subplt_kw=None,gridspec_kw=None,**fig_kw)

参数说明:sharex,sharey是看是否为共享x轴y轴

squeeze:布尔值，默认为True

subplot_kw:可选，字典类型，把字典关键字传递给add_subplot()来创建每个子图

gridspec_kw:可选，字典类型，把字典的关键字传递给GridSpec构造函数创建子图放在网格里(grid)。

**fig_kw:把详细的关键字参数传给figure()函数

```python
import matplotlib.pyplot as plt
import numpy as np

#创建一些测试数据  -- 图1
x = np.linspace(0,2*np.pi,400)
y = np.sin(x**2)

#创建一个画像和子图 --图2
fig ,ax = plt.subplots()
ax.plot(x,y)
ax.set_title('Simple plot')

#创建两个子图  --图3
f,(ax1,ax2) = plt.subplots(1,2,sharey=True)
ax1.plot(x,y)
ax2.set_title('Sharing Y axis')
ax2.scatter(x,y)

#创建四个字图  --图4
fig,axs = plt.subplots(2,2,subplot_kw=dict(projection="polar"))
axs[0,0].plot(x,y)
axs[1,1].scatter(x,y)

#共享x轴
plt.subplots(2,2,sharex = 'col')

#共享y轴
plt.subplots(2,2,sharey = 'row')

#共享x轴和y轴
plt.subplots(2,2,sharex = 'all',sharey = 'all')

#创建10张图，已经存在的则会删除
fig,ax = plt.subplots(num=10,clear=True)
plt.show
```
