```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.linspace(-1,1,50)  #生成从-1到1的50个点
y = 2*x+1
```

```python
plt.plot(x,y)
plt.show()
```

```python
x = np.linspace(-3,3,50)  #生成从-1到1的50个点
y1 = 2*x+1
y2 = x**2
```

```python
plt.figure(num=1,figsize=(8,5))
plt.plot(x,y1,color='red',linewidth=1,linestyle='--')
plt.figure(num=3,figsize=(4,5))
plt.plot(x,y2)
```
