# Matplotlib网格线
使用grid()设置网格线

语法格式：matplotlib.pyplot.grid(b=None,which='major',axis='both')

b:可选，默认为None，true为显示网格线，false为不显示，如果设置**kwargs参数就为true

which:可选，可选值有major,minor,both，默认为major,表示应用更改的网格线

axis:可选，设置显示哪个方向的网格线，可以取both，x，y

**kwargs:可选，设置网络样式，可以是color = 'r',linestyle = '-'和linewidth = 2

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([1,2,3,4])
y = np.array([1,4,9,16])
plt.title("RUNOOB grid() Test")
plt.xlabel("x-label")
plt.ylabel("y-label")

plt.plot(x,y)
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([1,2,3,4])
y = np.array([1,4,9,16])
plt.title("RUNOOB grid() Test")
plt.xlabel("x-label")
plt.ylabel("y-label")

plt.plot(x,y)
plt.grid(c = 'red',ls = '--',linewidth = 0.5)
plt.show()
```
