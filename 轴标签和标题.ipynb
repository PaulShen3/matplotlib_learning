# 轴标签和标题
xlabel和ylabel设置x轴和y轴的标签

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([1,2,3,4])
y = np.array([1,4,9,16])
plt.plot(x,y)
plt.xlabel("x-label")
plt.ylabel("y-label")
plt.show()
```

# 标题
用title()来设置标题

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([1,2,3,4])
y = np.array([1,4,9,16])
plt.plot(x,y)
plt.title("RUNOOB TEST TITLE")
plt.xlabel("x-label")
plt.ylabel("y-label")
plt.show()
```

# 标签定位
title(),xlabel()都可以使用loc参数设置为left,right,center,默认为center,ylabel()使用bottom和top以及center

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([1,2,3,4])
y = np.array([1,4,9,16])
plt.plot(x,y)
plt.title("RUNOOB TEST TITLE")
plt.xlabel("x-label",loc = 'left')
plt.ylabel("y-label",loc = 'top')
plt.show()
```
